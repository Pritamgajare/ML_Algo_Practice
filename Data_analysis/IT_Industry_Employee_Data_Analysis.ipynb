**IT Industry Employee Data Analysis**

```python
# Import necessary libraries
import pandas as pd

# Load the dataset
df_employee = pd.read_csv('employee_data.csv')

# Display basic information about the dataset
df_employee.info()
```

```python
# Display the first few rows of the dataset
df_employee.head()
```

```python
# Summary statistics for numerical columns
df_employee.describe()
```

```python
# Count of unique values in each column
df_employee.nunique()
```

```python
# Check for missing values in the dataset
df_employee.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot the distribution of gender
plt.figure(figsize=(8, 6))
sns.countplot(data=df_employee, x='Gender', palette='viridis')
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Plot the distribution of experience in years
plt.figure(figsize=(10, 6))
sns.histplot(df_employee['Experience (Years)'], kde=True, color='teal')
plt.title('Distribution of Experience (Years)')
plt.xlabel('Experience (Years)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the distribution of salaries
plt.figure(figsize=(10, 6))
sns.histplot(df_employee['Salary'], kde=True, color='orange')
plt.title('Distribution of Salary')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the count of each position
plt.figure(figsize=(12, 8))
sns.countplot(data=df_employee, x='Position', palette='magma', order=df_employee['Position'].value_counts().index)
plt.title('Count of Positions')
plt.xlabel('Position')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate average salary by position
avg_salary_by_position = df_employee.groupby('Position')['Salary'].mean().reset_index()

# Plot the average salary by position
plt.figure(figsize=(12, 8))
sns.barplot(data=avg_salary_by_position, x='Position', y='Salary', palette='coolwarm')
plt.title('Average Salary by Position')
plt.xlabel('Position')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate average salary by gender
avg_salary_by_gender = df_employee.groupby('Gender')['Salary'].mean().reset_index()

# Plot the average salary by gender
plt.figure(figsize=(8, 6))
sns.barplot(data=avg_salary_by_gender, x='Gender', y='Salary', palette='pastel')
plt.title('Average Salary by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Salary')
plt.show()
```

```python
# Scatter plot of Experience vs Salary
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_employee, x='Experience (Years)', y='Salary', hue='Gender', palette='Set1', alpha=0.7)
plt.title('Experience vs Salary')
plt.xlabel('Experience (Years)')
plt.ylabel('Salary')
plt.legend(title='Gender')
plt.show()
```

```python
# Compute the correlation matrix
corr = df_employee[['Experience (Years)', 'Salary']].corr()

# Plot the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Plot the gender distribution by position
plt.figure(figsize=(12, 8))
sns.countplot(data=df_employee, x='Position', hue='Gender', palette='dark')
plt.title('Gender Distribution by Position')
plt.xlabel('Position')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot the distribution of experience by gender
plt.figure(figsize=(10, 6))
sns.histplot(data=df_employee, x='Experience (Years)', hue='Gender', multiple='stack', palette='tab10')
plt.title('Experience Distribution by Gender')
plt.xlabel('Experience (Years)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot of salary by position
plt.figure(figsize=(12, 8))
sns.boxplot(data=df_employee, x='Position', y='Salary', palette='crest')
plt.title('Salary Distribution by Position')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plot of Experience vs Salary with a regression line
plt.figure(figsize=(10, 6))
sns.regplot(data=df_employee, x='Experience (Years)', y='Salary', scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Experience vs Salary with Regression Line')
plt.xlabel('Experience (Years)')
plt.ylabel('Salary')
plt.show()
```

```python
# Scatter plot of Experience vs Salary by Gender with regression lines
plt.figure(figsize=(12, 8))
sns.lmplot(data=df_employee, x='Experience (Years)', y='Salary', hue='Gender', palette='Set1', aspect=1.5, height=6)
plt.title('Experience vs Salary by Gender with Regression Lines')
plt.xlabel('Experience (Years)')
plt.ylabel('Salary')
plt.show()
```

```python
# Plot salary distribution by gender and position
plt.figure(figsize=(14, 8))
sns.boxplot(data=df_employee, x='Position', y='Salary', hue='Gender', palette='muted')
plt.title('Salary Distribution by Gender and Position')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.xticks(rotation=45)
plt.show()
```

```python
# Facet grid of Experience vs Salary by Position
g = sns.FacetGrid(df_employee, col='Position', col_wrap=3, height=4, aspect=1.5)
g.map_dataframe(sns.scatterplot, x='Experience (Years)', y='Salary', hue='Gender', palette='tab10', alpha=0.7)
g.add_legend()
g.set_axis_labels('Experience (Years)', 'Salary')
g.set_titles(col_template="{col_name}")
plt.show()
```
